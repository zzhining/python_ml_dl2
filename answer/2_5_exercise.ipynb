# 문제1
주어진 데이터프레임 df를 사용하여 과목별 평균 점수를 구하시오.

````python
import pandas as pd

data = {
    '이름': ['혜슬', '혜슬', '철수', '영희', '철수', '영희'],
    '과목': ['수학', '과학', '수학', '과학', '과학', '수학'],
    '점수': [85, 90, 78, 88, 95, 80]
}

df = pd.DataFrame(data)
````

<힌트>
```
groupby()
```

````python
# 코드 작성

# 과목별로 평균 점수 구하기
grouped = df.groupby('과목')['점수'].mean()
display(grouped)
````

# 문제2
주어진 데이터프레임 df를 사용하여 학생별, 과목별 평균 점수를 구하시오.

````python
import pandas as pd

data = {
    '이름': ['혜슬', '혜슬', '철수', '영희', '철수', '영희'],
    '과목': ['수학', '과학', '수학', '과학', '과학', '수학'],
    '점수': [85, 90, 78, 88, 95, 80]
}

df = pd.DataFrame(data)
````

<힌트>
```
pivot()
pivot_table()
crosstab()
```

예시1 – pivot(index, columns, values) 활용

````python
# 코드 작성

# 피벗 테이블로 변환
pivot_df = df.pivot(index='이름', columns='과목', values='점수')
display(pivot_df)
````

예시2 – pivot_table(index, columns, values, aggfunc) 활용

````python
# 피벗 테이블 생성 및 평균 계산
pivot_table_df = df.pivot_table(index='이름', columns='과목', values='점수', aggfunc='mean')
display(pivot_table_df)
````

예시3 - crosstab(index, columns, values, aggfunc) 활용

````python
# 과목과 이름에 따른 점수 합계
crosstab_df = pd.crosstab(df['이름'], df['과목'], values=df['점수'], aggfunc='sum')
display(crosstab_df)
````

# 문제6
다음 데이터프레임에 대해 정규화, 표준화한 값을 갖는 데이터 프레임을 각각 생성하여 출력하시오.

````python
import pandas as pd

# 샘플 데이터 생성
data = {'나이': [25, 30, 45, 50, 60],
        '연봉': [3500, 4000, 5000, 5500, 6500]}

df = pd.DataFrame(data)
````

<힌트>
```
MinMaxScaler()
StandardScaler()
```

정규화 결과

````python
# 코드 작성

from sklearn.preprocessing import MinMaxScaler

# Min-Max 정규화
min_max_scaler = MinMaxScaler()
df_minmax = pd.DataFrame(min_max_scaler.fit_transform(df), columns=df.columns)
print("Min-Max 정규화된 데이터:")
display(df_minmax)
````

````python
from sklearn.preprocessing import StandardScaler

# Z-점수 정규화 (표준화)
standard_scaler = StandardScaler()
df_standard = pd.DataFrame(standard_scaler.fit_transform(df), columns=df.columns)
print("Z-점수 정규화된 데이터:")
display(df_standard)
````

# 문제 7
사이킷런 패키지는 다양한 데이터셋을 제공한다. 대표적인 데이터 셋인 붓꽃 데이터셋(iris data)은 데이터 정제가 잘 되어있고, 머신러닝 학습 진행 시 성능이 좋아 머신러닝을 이제 막 시작하는 초보자들이 학습용으로 사용하는 대표적인 데이터 셋이다.
붓꽃 데이터셋은 꽃받침 길이와 너비, 꽃잎 길이와 너비의 4개의 특징 데이터에 따른 품종을 구분한 데이터로 총 150개의 값이 존재한다.

````python
import pandas as pd
from sklearn.datasets import load_iris

# 데이터 불러오기
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
````

1. 붓꽃 데이터셋(df)의 원본 데이터에 대해 값의 분포를 확인하고
2. `species` 열을 제외한 모든 열에 대해 정규화와 표준화를 각각 수행하여 데이터 분포를 확인하시오.

<힌트>
```
MinMaxScaler()
StandardScaler()
```

````python
# 코드 작성

import pandas as pd
from sklearn.datasets import load_iris

# 데이터 불러오기
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target

# 데이터 분포 확인
df.describe()
````

MinMaxScaler 수행 결과

````python
from sklearn.preprocessing import MinMaxScaler

# 정규화 수행을 위한 객체 생성
minmax_scaler = MinMaxScaler()

X = df.drop('species', axis=1)

# 데이터 변환을 위한 기준값 설정
minmax_scaler.fit(X)
# 정규화 수행
minmax_scaled_data = minmax_scaler.transform(X)

# 정규화 수행 결과를 포함하는 데이터 프레임 생성
minmax_scaled_df = pd.DataFrame(data=minmax_scaled_data, columns=iris.feature_names)
# 통계 분포 확인
minmax_scaled_df.describe()
````

StandardScaler 결과

````python
from sklearn.preprocessing import StandardScaler

# 표준화 수행을 위한 객체 생성
standard_scaler = StandardScaler()

# 데이터 변환을 위한 기준값 설정
standard_scaler.fit(X)

# 표준화 수행
standard_scaled_data = standard_scaler.transform(X)

# 표준화 수행 결과를 포함하는 데이터 프레임 생성
standard_scaled_df = pd.DataFrame(data=standard_scaled_data, columns=iris.feature_names)

# 통계 분포 확인
standard_scaled_df.describe()
````
